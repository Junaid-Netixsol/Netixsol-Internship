# Week 3 – Thursday: Advanced SQL Business Intelligence Challenge (Music Store Database)

## Goal

- Work with a new enterprise database and solve a complete business problem using SQL. Instead of writing independent queries, build a chain of analytical queries where each step depends on the previous one. Students should decide how to structure their solution using CTEs, window functions, subqueries, aggregations, and joins.

---

## Learning Topics

- Multi-level CTEs
- Chaining CTEs together
- Window Functions
- CASE WHEN
- Conditional Aggregation
- Ranking Functions
- Business KPIs
- Analytical Thinking
- SQL Query Organization

## Concept Check

- Why are multiple CTEs preferred over one large nested query?
- When would you use a window function instead of GROUP BY?
- Explain the difference between ROW_NUMBER(), RANK(), and DENSE_RANK().
- What is conditional aggregation?
- How does CASE WHEN improve analytical reporting?
- Why should SQL queries be broken into logical stages?
- What makes a SQL query maintainable?

## Implementation Task Scenario

- Analyze the Music Store database and solve the following business problems using SQL. Your solution should be organized as a reusable SQL pipeline, where each stage depends on the previous one.

---

### Task 1 — Build Customer Spending Profiles

First, create a reusable customer profile containing:

- Total amount spent
- Total invoices
- Total tracks purchased
- Number of unique genres purchased
- Number of unique artists purchased
- Number of purchase months
- Average invoice value

Requirements:

- Use one or more CTEs.
- Avoid repeating calculations.
- This dataset must be reused in later tasks.



---

### Task 2 — Customer Segmentation
- Using the customer profile from Task 1:
- Classify every customer into one of four segments:

- Platinum
- Gold
- Silver
- Bronze

- Your segmentation must consider multiple factors, not just spending.

  - For example:
    - Total spending
    - Purchase frequency
    - Genre diversity
    - Artist diversity
    - Design your own segmentation logic and justify it in the README.
    - Requirements

- Continue from Task 1.
- Use CASE WHEN.
- Do not rebuild calculations from scratch.



---

### Task 3 — Personalized Marketing Recommendation
- Using the customer segments created in Task 2:
- Determine the favorite genre of each customer.
- Then recommend one promotional campaign for each segment.
- Example:
    - Platinum → Early access to new releases
    - Gold → Album bundles
    - Silver → Genre discounts
    - Bronze → First purchase coupons
    - Support every recommendation with SQL results.
    - Requirements

- Continue from previous CTEs.
- Use ranking/window functions to determine favorite genres.



---

### Task 4 — Country Expansion Strategy
- Management wants to expand into new countries.
- Develop a Country Performance Score using multiple business metrics:

- Total Revenue
- Total Customers
- Average Revenue per Customer
- Average Invoice Value
- Number of Genres Purchased
- Customer Diversity
- Create your own weighted scoring formula and rank all countries.
- Finally recommend the top three countries for future expansion and explain your reasoning.
- Requirements

- Use multiple CTEs.
- Use window functions for ranking.
- Use calculated business metrics rather than simple totals.



---

### Task 5 — Executive SQL Report
- Using all previous work, build a final SQL report containing:

- Customer Segment Summary
- Revenue by Segment
- Top Customer in each Segment
- Top Genre in each Segment
- Best Performing Country
- Revenue Contribution by Country
- Top Employee by Revenue
- Top Artist by Revenue
- Top Album by Revenue
- This report should reuse the outputs from previous tasks instead of recalculating everything.
