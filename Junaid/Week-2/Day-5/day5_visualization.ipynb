```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../Day-3/pakistan_economic_indicators_2000_2025.csv")
df.head()
```

## Histogram

```python
plt.hist(df["gdp_usd_bn"], label="GDP", edgecolor="black")
plt.title("GDP of Pakistan")
plt.xlabel("GDP (USD) in Billions")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

- histogram is used to see spread and frequency of continuous values. you only need to provide 1 column, and that column will be on the x axis, frequeny will be on the right axis. some things to keep in mind:
- the frequencies all summed up should equal 26 here (we got 26 rows)
- each bar is a range, not an exact value (for instance the first bar is 100-130 ish)

- so, pakistan has had a gdp of 100-130 billion about 5 times over the past 26 years, same for a gdp of 250-280 billion as well

## Bar Chart

```python
plt.bar(df["year"], df["gdp_usd_bn"], label="GDP")
plt.title("GDP of Pakistan Over Time")
plt.xlabel("Year")
plt.ylabel("GDP (USD) in Billions")
plt.legend()
plt.show()
```

- barcharts are for simple comparisons like this, as in gdp over the years, so two columns are involved
- interpretation should be quite simple, pakistan has had a steadily increasing GDP over the years, the only place that seems unusual is around 2019-2020. That was probably because of Covid

## Box Plot

```python
plt.boxplot(df["exports_gdp_pct"])
plt.title("Median and Spread of How much Exports are part of the GDP")
plt.ylabel("% of GDP")
plt.show()
```

- boxplot shows what range the majority of data is in, how much spread out it is, and it also catches outliers
- so generally speaking, exports are about 8.5 to 10.5 percent of the total GDP of the year

## Correlation Heat Map

```python
corr = df.corr(numeric_only=True)
sns.heatmap(
    corr, cmap="coolwarm", 
)
plt.title("Correlation Between Different Variables")
plt.show()
```

- heatmaps are used to find patterns across columns. the cell in the heatmap where the row and column intersect is generally the darkest, it just means the exact same columns are intersecting so correlation is 1.0. 
- From the above chart, its obvious that population has a very high correlation with remittances.
- also, mobile subscription per 100 people has a very high correlation to population.
- pkr per usd increases significantly as population grows
- inflation is kinda directly proportional with exports
