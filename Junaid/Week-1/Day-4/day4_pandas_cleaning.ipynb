## Handling Missing Values
---

- its explicitly stated to use the dataset from day 3 but that one has no missing values or duplicates or wrongly type casted columns, so I will just focus on learning each step generally


- I will go through diff methods of finds and printing out missing values and also go through dealing with them (drop them or fill them up). Will put tny comments in each cell

```python
# to find exact empty cells
df.isna()
```

```python
# find total missing cells in each column
df.isna().sum()
```

```python
# for percentage of missing values per column
df.isna().mean() * 100
```

```python
# return just one boolean per row
df.isna().any()
```

### Dealing with missing values

```python
df.shape
```

```python
# drop any row that has even one missing value
df.dropna()
```

```python
df.shape
```

```python
# drop any column that has even one missing value   
df.dropna(axis=1)
```

```python
# drop rows where you have 3 missing values at least
df.dropna(thresh=3)
# just do axis=1 for columns
```

```python
# only check specific columns for missing values
df.dropna(subset=["gdp_growth_category", "decade"])
```

```python
# Instead of deleting, you can also just try to fill in the missing values btw
df.fillna("NA")
```

```python
# you can just fill the missing value with the mean, median, or mode of the respective column that missing value is in
df["gdp_usd_bn"].fillna(df["gdp_usd_bn"].mean())
```

```python
# fill the missing value with the value of that same column from the previous row, this is mostly used in time series data
df.ffill()
```

```python
# the opposite of ffill
df.bfill()
```

## Feature Engineering (make new derived columns)

- feature engineering is inventing new data from existing data, like multiple length of width to get area etc. You have 2 simple ways of doing it, just do direct maths on columns and get a new column or make a custom function and use .apply to use it

```python
# Simple addition
df['export_import_pct'] = df['exports_gdp_pct'] + df['imports_gdp_pct']
# Combining text columns
# going to kep this commented out since it will throw an error
# df['full_name'] = df['first_name'] + ' ' + df['last_name']
```

```python
def double_gdp(gdp):
    return gdp * 2

df["double_gdp"] = df["gdp_usd_bn"].apply(double_gdp)
summary_df = df.groupby('year').agg({
    'gdp_usd_bn': ['mean', 'max'],
})
summary_df
```

```python
# pivot_table is a bit complicated. 
# the first arg it takes in is index, this will become the vertical axis at the start of each row
# the second param is values, which is what you wanna calc, by default, it will just fine mean and nothing else
df1 = df.pivot_table(index="key_events", values="gdp_usd_bn")
df1
```

```python
# if you don't want mean and want something else, you can use aggfunc
df1 = df.pivot_table(index="key_events", values="gdp_usd_bn", aggfunc="sum")
df1
```

## Merging dfs and mapping a column to labels

- mapping a column to a label is simply making a mapper (just a dictionary) and in that dict, you map the values of a column to a certain label, then you use .map on the df with that column selected
- merging tables is more or less the same as sql joins, you have a foreign key, and you have the type of join, inner or left or right or outer etc

```python
mapper = {0: "not_under_IMF_plan", 1: "under_IMF_plan"}

df["imf_program_active"] = df["imf_program_active"].map(mapper)
df.head()
```

```python
# we can basically do sql joins using merge
df2 = {"year": 2025, "dummy_column": 400}
df2 = pd.DataFrame(df2, index=[0])

merged_df = pd.merge(df, df2, on="year", how="left")
```
